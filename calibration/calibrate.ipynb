```python
import matplotlib.pyplot as plt
import numpy as np
import json
import math
```

```python
with open('result.json') as data_file:    
    calibrate_matrix = json.load(data_file)
```

```python
fx = []
fy = []
cx = []
cy = []
for i in range(0,len(calibrate_matrix)):
    fx.append(calibrate_matrix[i][0][0])
    cx.append(calibrate_matrix[i][0][2])
    fy.append(calibrate_matrix[i][1][1])
    cy.append(calibrate_matrix[i][1][2])
```

```python
print(np.mean(cx[30:]))
print(np.mean(cy[30:]))
```

```python
x = list(range(10,65))
plt.figure(figsize=(16,9))
plt.plot(x,fx,'r',label = 'fx')
plt.plot(x,fy, 'b',label = 'fy')
plt.title('focal lengths change with number of images', fontsize=24)
plt.xlabel('number of images used', fontsize=24)
plt.ylabel('coordinate in pixels', fontsize=24)
plt.legend()
plt.savefig('focal.jpg')
plt.show()
```
