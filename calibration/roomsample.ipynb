```python
import numpy as np
import json
import math
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
with open('../output/2017-05-29-12-02-22/result.json') as data_file:    
    data1 = json.load(data_file)
with open('../output/2017-05-29-12-43-09/result.json') as data_file:
    data2 = json.load(data_file)
with open('../output/2017-05-29-12-44-54/result.json') as data_file:
    data3 = json.load(data_file)
with open('../output/2017-05-29-12-45-33/result.json') as data_file:
    data4 = json.load(data_file)
with open('../output/2017-05-29-12-47-28/result.json') as data_file:
    data5 = json.load(data_file)
with open('../output/2017-05-29-12-55-35/result.json') as data_file:
    data6 = json.load(data_file)
with open('../output/2017-05-29-12-58-22/result.json') as data_file:
    data7 = json.load(data_file)
with open('../output/2017-05-29-13-03-15/result.json') as data_file:
    data8 = json.load(data_file)   
with open('../output/2017-05-29-13-05-58/result.json') as data_file:
    data9 = json.load(data_file) 
with open('../output/2017-05-29-13-15-09/result.json') as data_file:
    data10 = json.load(data_file)

data = [data1,data2,data3,data4,data5,data6,data7,data8,data9,data10]
```

```python
location = []
for element in data:
    location.append((element[0][0]['ground_truth'], element[0][0]['camera_location']))
    
d = []
error = []
for element in location:
    d.append(math.sqrt(math.pow(element[0][0],2)+math.pow(element[0][1],2)+math.pow(element[0][2],2)))
    error.append(math.sqrt(math.pow(element[0][0]-element[1][0],2)+math.pow(element[0][1]-element[1][1],2)+math.pow(element[0][2]-element[1][2],2)))
    
ber = []
for i in range(0,len(d)):
    ber.append(error[i]/d[i])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(d,ber)
plt.title('Relation between error ratio and distance')
plt.xlabel('Distance to origin point [m]')
plt.ylabel('error ratio')
plt.xlim(2,6.5)
plt.ylim(0,0.1)
ax.annotate('error=0.015',(d[-1],ber[-1]))
plt.show()
```
